## Paths and executables

```bash
conda activate gapseq
gapseq -v
conda deactivate
```

```bash
# Paths to input genome data
pangenome="01-pangenome/clostridiales.faa"

# Path to results folder
workFolder="02-pan_model"
mkdir -p $workFolder

# root folder of this analysis
root=$(pwd)
```

## Parameters

```bash
# Setting non-default gapseq parameters
gs_blast_sens=150
n_cores=14
```

## Clean up any character gapseq does not recognise

```bash
cat $pangenome | tr -d "’" > tmp
mv tmp $pangenome
```

## Generating models

### Find metabolic pathways

```bash
conda activate gapseq
```

```bash
# Find metabolic pathways from the genome
gapseq find \
-p all -v 1 -y -j -O \
-b $gs_blast_sens \
-f $workFolder \
-K $n_cores \
$pangenome \
> $workFolder/log.log \
2> $workFolder/log.err
```

```bash
conda deactivate
```
