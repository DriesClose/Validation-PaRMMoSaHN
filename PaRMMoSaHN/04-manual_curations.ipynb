This is a placeholder notebook for the manual curation part of the pipeline. As the name implies, this is a manual step but can be skipped if you prefer.

There is no determined file structure for this part but I recommend you do implement one yourself. Keep at least all the resulting GSMMs separate in a subfolder so you can refer to it in the later pipeline parts.

Personally, I use the following directory structure:

- a folder `analyses` for a Jupyter notebook for each genome, in which I iteratively check the Biolog accuracy of each model while manual curating. You can take these notebooks as a start for your analyses if you want to.
- a folder `models` which contains the manually curated models.
- a folder `scripts` containing the Python scripts that execute all manual curation steps for each model.
- some Excel files in the root of this `04-` folder with some logging and metadata about the curations.
